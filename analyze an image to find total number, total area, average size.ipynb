###LAB-3: Program to analyze and image to find total number, total area, average size of elements

```python
import cv2
import numpy as np
from google.colab.patches import cv2_imshow  # Import cv2_imshow for displaying images in Colab

# Load the image
image = cv2.imread('/content/gettyimages-1152347304-612x612.jpg')

# Display the original image
print("Original Image:")
cv2_imshow(image)  # Display the original image

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply binary thresholding to separate objects from the background
_, binary = cv2.threshold(gray, 128, 255, cv2.THRESH_BINARY_INV)

# Find contours (which correspond to elements)
contours, _ = cv2.findContours(binary, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# Total number of elements
num_elements = len(contours)

# Calculate the total area and average size of elements
total_area = 0

for contour in contours:
    area = cv2.contourArea(contour)
    total_area += area

# Average size of elements
if num_elements > 0:
    average_size = total_area / num_elements
else:
    average_size = 0

# Display results
print(f"Total number of elements: {num_elements}")
print(f"Total area of elements: {total_area}")
print(f"Average size of elements: {average_size}")

# Visualize contours
output_image = cv2.drawContours(image.copy(), contours, -1, (0, 255, 0), 2)

# Display the image with contours
print("Image with Contours:")
cv2_imshow(output_image)  # Use cv2_imshow to display the image with contours
```
