###Program to implement contrast stretching.

**Contrast Streching**  also known as normalization, is a simple image enhancement technique used to improve the contrast of an image by expanding its intensity range. The goal of contrast stretching is to make the dark areas darker and the bright areas brighter, thereby enhancing the visibility of details in an image that may otherwise be lost in low-contrast regions.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/dull_image.jpeg'  # Path to your image
image = cv2.imread(image_path, 0)  # Load in grayscale

# Make the original image dull by reducing contrast
dull_image = cv2.convertScaleAbs(image, alpha=0.7, beta=0)  # Reduce contrast (alpha < 1 for dull effect)

# Perform Contrast Stretching
min_val = np.min(dull_image)
max_val = np.max(dull_image)

# Stretching contrast
stretched_image = ((dull_image - min_val) / (max_val - min_val)) * 255
stretched_image = np.clip(stretched_image, 0, 255).astype(np.uint8)

# Display the dull and contrast-stretched images side by side
plt.figure(figsize=(8,6))  # Adjust the figure size for side-by-side display

# Dull Image
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.title("Dull Image")
plt.imshow(dull_image, cmap='gray')
plt.axis('off')

# Contrast-Stretched Image
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.title("Contrast Stretched Image")
plt.imshow(stretched_image, cmap='gray')
plt.axis('off')

# Show both images
plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = '/content/800wm.jpeg'  # Update with the actual path of the image
image = cv2.imread(image_path, 0)  # Load the image in grayscale

# Make the original image dull by reducing contrast
dull_image = cv2.convertScaleAbs(image, alpha=0.7, beta=0)

# Perform Contrast Stretching
min_val = np.min(image)  # Minimum pixel value in the image
max_val = np.max(image)  # Maximum pixel value in the image

# Apply the contrast stretching formula
stretched_image = ((image - min_val) / (max_val - min_val)) * 255
stretched_image = np.clip(stretched_image, 0, 255).astype(np.uint8)  # Ensure pixel values stay between 0 and 255

# Apply a color map to the contrast-stretched image to make it colorful
colorful_image = cv2.applyColorMap(stretched_image, cv2.COLORMAP_JET)

# Reduce the intensity of the colorful image by scaling down the pixel values
reduced_intensity_image = cv2.convertScaleAbs(colorful_image, alpha=0.5, beta=0)  # Adjust alpha to reduce intensity

# Display the original and reduced-intensity colorful contrast-stretched images side by side
plt.figure(figsize=(8, 6))  # Adjust figure size for better visibility

# Original Grayscale Image
plt.subplot(1, 2, 1)  # 1 row, 2 columns, first subplot
plt.title("Original Grayscale Image")
plt.imshow(image, cmap='gray')
plt.axis('off')

# Reduced Intensity Colorful Contrast-Stretched Image
plt.subplot(1, 2, 2)  # 1 row, 2 columns, second subplot
plt.title("Reduced Intensity Colorful Image")
plt.imshow(cv2.cvtColor(reduced_intensity_image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Show both images
plt.show()

# Save the reduced intensity colorful contrast-stretched image
cv2.imwrite('reduced_intensity_colorful_contrast_stretched_image.jpeg', reduced_intensity_image)
```
